# OLIVES Feasibility Audit

**Goal:** Determine whether OLIVES has the temporal structure to support disease-dynamics modeling.  
**Decision gate:** ≥4 visits per eye → latent ODE viable. 2-3 visits → simpler temporal baselines.  
**No model code here — inspection and decision only.**

Dataset: OLIVES (Zenodo DOI: 10.5281/zenodo.7105232)  
Raw data: `../data/raw/olives/`

```python
import os
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from collections import defaultdict

DATA_DIR = Path('../data/raw/olives')
FIGURES_DIR = Path('../results/figures')
FIGURES_DIR.mkdir(parents=True, exist_ok=True)
```

## Step 1 — Unzip Labels

```python
labels_zip = DATA_DIR / 'OLIVES_Dataset_Labels.zip'
labels_dir = DATA_DIR / 'labels'

if not labels_dir.exists():
    print('Unzipping labels...')
    with zipfile.ZipFile(labels_zip) as zf:
        zf.extractall(labels_dir)
    print('Done.')
else:
    print('Labels already unzipped.')

# List what's in the labels dir
for p in sorted(labels_dir.rglob('*')):
    print(p.relative_to(labels_dir))
```

## Step 2 — File/Folder Structure

Inspect how eyes, visits, fundus, and OCT are organized in the main zip (once downloaded and unzipped).

```python
# Peek at the top-level structure of OLIVES.zip without fully extracting
olives_zip = DATA_DIR / 'OLIVES.zip'

if olives_zip.exists():
    with zipfile.ZipFile(olives_zip) as zf:
        names = zf.namelist()
    print(f'Total entries in zip: {len(names)}')
    # Show top-level dirs
    top_level = sorted(set(n.split('/')[0] for n in names))
    print('Top-level entries:')
    for t in top_level:
        print(' ', t)
    # Show first 50 entries as a sample
    print('\nFirst 50 entries:')
    for n in names[:50]:
        print(' ', n)
else:
    print('OLIVES.zip not yet downloaded — run this cell once the download completes.')
```

## Step 3 — Load Clinical Labels & Build Per-Eye Visit Table
